```python
import numpy as np
from scipy.special import sph_harm
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 
# m
# l
# px
# tz
def orbital(m, l, p, t):
    r = np.abs(sph_harm(m, l, p, t))
    x = r*np.sin(t)*np.cos(p)
    y = r*np.sin(t)*np.sin(p)
    z = r*np.cos(t)
    #r1 = np.sqrt(21.0/32.0/np.pi)*(x*(5.0*z*z-x*x-y*y-z*z))/(x*x+y*y+z*z)
    #print()
    return x, y, z

# FigureAxes
fig = plt.figure(figsize = (8, 6))
ax = fig.add_subplot(111, projection='3d')

# 
ax.set_xlabel("x", size = 16)
ax.set_ylabel("y", size = 16)
ax.set_zlabel("z", size = 16)

# 
ax.view_init(45, 45)

# theta,phi
t = np.linspace(0, np.pi, 10)
p = np.linspace(0, 2*np.pi, 10)

# 
tt, pp = np.meshgrid(t, p)

# l=3,m=1
x, y, z = orbital(1, 3, pp, tt)

# 
ax.plot_surface(x, y, z, color="darkorange",edgecolor="black", linewidth=0.4)
```
