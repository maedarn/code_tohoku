```python
# coding: UTF-8
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import h5py
import csv
import math
import pandas as pd
from matplotlib.backends.backend_pdf import PdfPages
import matplotlib.animation as animation
from matplotlib.animation import ArtistAnimation
from sympy import *
from sympy.codegen.cfunctions import log10
    
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

gam=5.0/3.0

r1=np.linspace( -10, 10, 100)
r2=(1.0-gam/3.0)*r1+2.0/27.0*(r1**3.0+(r1**2.0-3.0)**(3.0/2.0))
r3=(1.0-gam/3.0)*r1+2.0/27.0*(r1**3.0-(r1**2.0-3.0)**(3.0/2.0))



plt.plot(r1, r2)
plt.plot(r1, r3)

#ax.set_yscale('log')
#ax.set_xscale('log')

plt.xlim(-10, 10)
plt.ylim(-10, 10)

ax.set_aspect("equal")

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

gam=5.0/3.0
alpha=0.1
beta=100.0

x=np.linspace(0, 10, 100)
u=-gam*0.5/beta*(beta*x*x+x+(alpha-1.0)/gam)



plt.plot(x, u)

#ax.set_yscale('log')
#ax.set_xscale('log')

plt.xlim(0, 0.1)
plt.ylim(-0.1, 0.1)

#ax.set_aspect("equal")

plt.show()
```
